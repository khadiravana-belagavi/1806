```
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Question 8.1. 

Write a function gammaTransform that implements $s = r^{Î³}$ on a grayscale image and vary $\gamma$.

```
def gammaTransform(img,gamma):
  width,height = img.shape
  img_min = np.min(img)
  img_max = np.max(img)
  bandwidth = img_max - img_min # Clearly this cannot be zero for this particular image

  new_img = np.zeros((width,height))
  for i in range(width):
    for j in range(height):
      pix = img[i,j]
      new_pix = round(bandwidth*pow((pix/bandwidth),gamma))
      new_img[i,j] = new_pix
  return new_img
  
```

```
img = cv.imread('/content/drive/My Drive/DIP-Assignment1/assignment-1-khadiravana-belagavi/images/stealth.png')
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```
plt.imshow(img,cmap='gray')
```

```
transformed_img = gammaTransform(img,0.425) # 
```
